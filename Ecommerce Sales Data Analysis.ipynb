## Converting All the CSV files into DataFrame

```python
CustomerData = pd.read_csv('./DATA/olist_customers_dataset.csv')
OrderData = pd.read_csv('./DATA/olist_orders_dataset.csv')
OrderItemsData = pd.read_csv('./DATA/olist_order_items_dataset.csv')
OrderPaymentData = pd.read_csv('./DATA/olist_order_payments_dataset.csv')
OrderReviewData = pd.read_csv('./DATA/olist_order_reviews_dataset.csv')
ProductData = pd.read_csv('./DATA/olist_products_dataset.csv')
SellerData = pd.read_csv('./DATA/olist_sellers_dataset.csv')
CategoryTranslationData = pd.read_csv('./DATA/product_category_name_translation.csv')
```

## All the Data Frames to be merged into Single DataFrame.

- **Here `merge` function used to merge different dataframes based on the common column in 2 dataframes**

```python
FinalData = pd.merge(OrderItemsData,ProductData,on='product_id').merge(CategoryTranslationData,on='product_category_name').merge(SellerData,on='seller_id').merge(OrderReviewData,on='order_id').merge(OrderData,on='order_id').merge(OrderPaymentData,on='order_id').merge(CustomerData,on='customer_id')

FinalData.head()
```

## Exporting DataFrame to CSV file for futher analysis.

```python
FinalData.to_csv('finalData.csv')
```
